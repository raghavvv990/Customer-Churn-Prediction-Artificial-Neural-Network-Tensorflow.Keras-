```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder , OneHotEncoder , StandardScaler
from sklearn.model_selection import train_test_split
 
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df = df.drop(['RowNumber','CustomerId','Surname'],axis=1)
df
```

```python
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)

ohe_geo = ohe.fit_transform(df[['Geography']])

ohe_geo = pd.DataFrame(
    ohe_geo,
    columns=ohe.get_feature_names_out(['Geography']),
    index=df.index
)

df = pd.concat([df.drop('Geography', axis=1), ohe_geo], axis=1)
df
```

```python
import pickle
```

```python
with open('le.pkl','wb') as file:
    pickle.dump(le,file)

with open('ohe.pkl','wb') as file:
    pickle.dump(ohe,file)
```

```python
X = df.drop(['Exited'],axis=1)
y = df[["Exited"]]
```

```python
X
```

```python
y
```

```python
X_train , X_test , y_train, y_test = train_test_split(X,y,test_size= .25 ,random_state=42)
```

```python
ss = StandardScaler()
```

```python
X_train = ss.fit_transform(X_train)
```

```python
X_test = ss.transform(X_test)
```

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import TensorBoard , EarlyStopping
import datetime
```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
import tensorflow
opt =  tensorflow.keras.optimizers.Adam(learning_rate=.01)
loss = tensorflow.keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer=opt,loss = loss,metrics=['accuracy'])
```

```python
log_dir = 'log/fit'+ datetime.datetime.now().strftime('%Y%m%d-%H%M%S')
```

```python
log_dir
```

```python
tensorflow_callback = TensorBoard(log_dir=log_dir,histogram_freq=1)
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
history = model.fit(X_train,y_train, validation_data=(X_test,y_test),epochs=100, 
                    callbacks=[tensorflow_callback,early_stopping_callback])
```

```python
model.save('model.h5')
```
